```python
import cv2
from google.colab.patches import cv2_imshow

# Specify the file path
image_path = '/content/drive/MyDrive/imagr.jpg'

# Read the image
image = cv2.imread(image_path)

# Check if the image is loaded successfully
if image is not None:
    # Display the image
    cv2_imshow(image)
else:
    print("Failed to load the image.")

```

```python
#Coverting the color from BGR to Gray
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
#Display the converted image
cv2_imshow(gray_image)
```

```python
#Invert the image
inverted = 255- gray_image
```

```python
#check the inverted image
cv2_imshow(inverted)
```

```python
#Make blur to inverted image using higest filter size
blur_image = cv2.GaussianBlur(inverted, (111, 111), 0)




```

```python

#Invert the blur image
inverted_blur  = 255 - blur_image
```

```python
#Display the inverted blured image
cv2_imshow(inverted_blur)
```

```python
#Generate a sketch by dividing the greyscale image to the blur image
sketch = cv2.divide(gray_image, inverted_blur, scale = 256.0)


#Diplay the final result
cv2_imshow(sketch)
```

```python
#Save the Result
cv2.imwrite('Sketch.jpg', sketch)
```
